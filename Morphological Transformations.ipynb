```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread("Dark.jpg", 0)
plt.imshow(image, cmap="gray")
plt.show()
```

```python
_,bi_img =cv.threshold(image, 127, 255, cv.THRESH_BINARY)
plt.imshow(bi_img, cmap = "gray")
plt.show()
```

```python
_, mask = cv.threshold(image, 127, 255, cv.THRESH_BINARY_INV)
plt.imshow(mask, cmap = "gray")
plt.show()
```

DILATION

```python
kernel = np.ones((2,2), np.uint8)
dilation = cv.dilate(mask, kernel, iterations = 2)
plt.imshow(dilation, cmap ="gray")
```

EROSION

```python
kernel = np.ones((2,2), np.uint8)
erosion = cv.erode(mask, kernel, iterations = 5)
plt.imshow(erosion, cmap ="gray")
```

OPENING

```python
opening = cv.morphologyEx(mask, cv.MORPH_OPEN, kernel)
plt.imshow(opening, cmap = "gray")
```

CLOSING

```python
closing = cv.morphologyEx(mask, cv.MORPH_CLOSE, kernel)
plt.imshow(closing, cmap = "gray")

# morphological gradient -  difference between erosion and dilation
# morphological top hat - difference between the opening and original image.
```
