```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("light.jpg", 0)
plt.imshow(image, cmap="gray")
```

```python
laplacian = cv2.Laplacian(image, cv2.CV_64F, ksize=1)
laplacian = np.uint8(np.abs(laplacian))
plt.imshow(laplacian, cmap="gray")
```

```python
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1)
sobelx = np.uint8(np.absolute(sobelx))
sobely = np.uint8(np.absolute(sobely))
plt.imshow(sobelx, cmap="gray")
```

```python
plt.imshow(sobely, cmap="gray")
```

```python
sobelcombine = cv2.bitwise_or(sobelx, sobely)
plt.imshow(sobelcombine, cmap="gray")
```

```python
def nothing(x):
    print(x)
canny = cv2.Canny(image, 100, 200)
cv2.namedWindow('Canny')
#gives lesser noise in canny than laplacian and sobel
cv2.createTrackbar('T1', 'Canny', 0, 255, nothing )
cv2.createTrackbar('T2', 'Canny', 0, 255, nothing )
cv2.imshow('Canny', canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
